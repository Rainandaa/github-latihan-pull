```python
#Read Simpan ke Variabel
readFile = SeqIO.read('./sequence.fasta', 'fasta')
seqF = readFile.seq
print(seqF)
```

```python
#Ambil DNA Pertama Sampai ke Stop Kodon (Transcribe, Translate)
mRNA = seqF.transcribe()
protein = seqF.translate()
print(protein)

#Cara 1 
protein1 = seqF.translate(to_stop=True)
print(protein1)

#Cara 2
sliced = protein.split('*')
print(sliced)
print(len(sliced)) #Untuk memberitahu seberapa besar ukuran arraynya
```

```python
#Pandas
import pandas as pd
#print(pd.array(sliced)) Akan Menampilkan Array
print(pd.DataFrame(sliced))
```

```python
#Declare Your Sequence
seq1 = seqF[:round(len(seqF)/2)] #Akan Ambil Setengah Sequence Awal sampai tengah Sequence
seq2 = seqF[round(len(seqF)/2)] #Akan Ambil Setengah Sequence Tengah sampai tengah Akhir
print(seq1)
print(seq2)
```

```python
#Protein Explanatory Data Analysis

#Membuat perbandingan antara GC Conteng di seq1 dengan seq2
from Bio.SeqUtils import GC
gc1 = GC(seq1)
gc2 = GC(seq2)
print(gc1,gc2)

#Perbandingan Melting Point seq1 dengan seq2
from Bio.SeqUtils import MeltingTemp as MT
mt1 = MT.Tm_NN(seq1)
mt2 = MT.Tm_NN(seq2)

#Diberi Plotting
from matplotlib import pyplot as plt
_,axs = plt.subplots(1,2)
axs[0].bar(["GC1", "GC2"],[gc1,gc2])
axs[1].bar(["mt1", "mt2"],[mt1,mt2])
plt.show()
```
